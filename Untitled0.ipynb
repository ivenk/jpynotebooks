```
path = untar_data(URLs.MNIST); path
URLs.MNIST
```

```
path.ls()
```

```
path_training = path/'training'
path_training.ls() [:3]
```

```
np.random.seed(2)
reg = r"[0-9]$"

trans = get_transforms(do_flip=False)
data = ImageDataBunch.from_folder(path, 'training', ds_tfms=trans, valid_pct=0.2)
```

```
data.show_batch(rows=5, figsize=(10,10))
```

```
learner = cnn_learner(data, models.resnet18, metrics=accuracy) 
```

```
learner.fit(2)
```

```
learner.save('1_resnet18')
```

```
intrp = ClassificationInterpretation.from_learner(learner)
```

```
intrp.plot_top_losses(10, heatmap=False)
```

```
intrp.plot_confusion_matrix()
```
