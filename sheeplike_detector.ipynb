Now seed the randomizer and create a imagedatabunch from the folder structure. We do not have the entire folder structure set up (imagenet style) so we need to set the location for the training data and generate the validation set using valid_pct=0.2.

```
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
                                  ds_tfms=get_transforms(),
                                  size=224, num_workers=4).normalize(imagenet_stats)
```

## Create model

```
learner = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```
learner.fit_one_cycle(1)
```

```
learner.save('stage-1')
```
